```python
import pickle

import numpy as np
```

```python
pickle_file = 'data/sensor_graph/adj_mx.pkl'
with open(pickle_file, 'rb') as f:
    pickle_data = pickle.load(f, encoding='latin1')
```

```python
sensor_ids, sensor_id_to_ind, adj_mx = pickle_data
```

```python
with open(r'E:\develop\gnn_cargo\embedding\id_label_set.pkl', 'rb') as f:
    id_label_set = pickle.load(f)
    
```

```python
adj_mx = np.zeros((len(id_label_set), len(id_label_set)))
```

```python
id_label_set
```

```python
#id_label_set转list
id_list = []
label_list = []
for i in id_label_set:
    id_list.append(i)
    label_list.append(id_label_set[i])
    
```

```python
for i in range(len(id_list)):
    for j in range(len(id_list)):
        if label_list[i] == label_list[j]:
            adj_mx[i][j] = 1
```

```python
adj_mx
```

```python
for i in range(len(id_list)):
    for j in range(len(id_list)):
        row1,col1 = id_list[i].split('_')
        row2,col2 = id_list[j].split('_')
        #判断是否相邻
        if abs(int(row1)-int(row2)) == 1 and abs(int(col1)-int(col2)) == 0:
            adj_mx[i][j] = 1
        elif abs(int(row1)-int(row2)) == 0 and abs(int(col1)-int(col2)) == 1:
            adj_mx[i][j] = 1
            
        
```

```python
adj_mx
```

```python
#保存
with open(r'E:\develop\gnn_cargo\embedding\adj_mx_test.pkl', 'wb') as f:
    pickle.dump(adj_mx, f, pickle.HIGHEST_PROTOCOL)
```
